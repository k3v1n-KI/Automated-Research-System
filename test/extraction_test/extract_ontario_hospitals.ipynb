```python
from pathlib import Path
import json
import re
from nodes.base import BaseNode
from openai import AsyncOpenAI


async def run_crawler():
    browser_config = BrowserConfig(verbose=True, headless=True)

    run_config = CrawlerRunConfig(
        # Content processing
        process_iframes=True,
        remove_overlay_elements=True,

        # Cache control
        cache_mode=CacheMode.BYPASS  # Bypass cache to ensure fresh content
    )
    asynchronous_crawler = AsyncWebCrawler(config=browser_config)
    async with asynchronous_crawler as crawler:
        crawl_result = await crawler.arun(url, run_config=run_config)
        if crawl_result.success:
            scrapes = crawl_result.markdown
        else:
            print(f"Crawl failed: {crawl_result.error_message}")
await run_crawler()

# Temporary extract node for this notebook: writes output per chunk
class NotebookCrawl4AIExtractNode(BaseNode):
    def __init__(self, char_limit: int = 12000):
        super().__init__()
        self.char_limit = char_limit
        api_key = os.getenv("OPENAI_API_KEY")
        if not api_key:
            raise ValueError("OPENAI_API_KEY not set in environment")
        self.client = AsyncOpenAI(api_key=api_key)
        self.model = os.getenv("OPENAI_MODEL", "gpt-5-mini")

    async def execute(self, state, progress):
        scraped_items = state.get("scraped_content", [])
        initial_prompt = state.get("initial_prompt", "")
        columns = state.get("columns") or state.get("column_specs") or []
        column_names = [col["name"] for col in columns if isinstance(col, dict) and col.get("name")] if columns else []
        column_constraint = ""
        if column_names:
            column_constraint = (
                f"\n\nIMPORTANT: Extract ONLY these columns in your JSON output: {', '.join(column_names)}."
                " Do not include any other fields."
            )

        output_dir = Path.cwd() / "test" / "extraction_test"
        if not output_dir.exists():
            output_dir = Path.cwd()
        output_path = output_dir / "extract_chunks.jsonl"
        if output_path.exists():
            output_path.unlink()

        extracted_items = []
        for idx, item in enumerate(scraped_items):
            url = item.get("url", "")
            markdown_text = item.get("text", "")
            if not markdown_text.strip():
                continue
            chunks = self._chunk_markdown(markdown_text)
            for chunk_index, chunk in enumerate(chunks, 1):
                if not chunk.strip():
                    continue
                try:
                    response = await self.client.chat.completions.create(
                        model=self.model,
                        messages=[
                            {
                                "role": "system",
                                "content": (
                                    "You are a data extraction expert. Extract structured information based on the task:\n\n"
                                    f"{initial_prompt}{column_constraint}\n\n"
                                    "Extract only valid, complete entries. Return a JSON array of objects."
                                ),
                            },
                            {
                                "role": "user",
                                "content": (
                                    f"Extract data from this content chunk ({chunk_index}/{len(chunks)}):\n\n"
                                    f"{chunk}\n\nURL: {url}"
                                ),
                            },
                        ],
                        max_completion_tokens=4000,
                    )
                    content = response.choices[0].message.content
                    json_match = re.search(r"\[.*\]", content, re.DOTALL)
                    records = []
                    if json_match:
                        data = json.loads(json_match.group())
                        if isinstance(data, list):
                            for record in data:
                                record["source_url"] = url
                                extracted_items.append(record)
                                records.append(record)
                    # Write per-chunk output
                    with open(output_path, "a", encoding="utf-8") as f:
                        f.write(json.dumps({
                            "url": url,
                            "chunk_index": chunk_index,
                            "records": records
                        }) + "\n")
                except Exception as e:
                    print(f"⚠️  Extraction error for {url}: {e}")
        state["extracted_items"] = extracted_items
        print("Chunk output file:", output_path)
        return state

    def _chunk_markdown(self, text: str):
        blocks = self._split_by_headers(text)
        if not blocks:
            return self._fixed_chunks(text)
        chunks = []
        bucket = []
        bucket_len = 0
        for block in blocks:
            block_len = len(block)
            if block_len > self.char_limit:
                if bucket:
                    chunks.append("\n".join(bucket))
                    bucket = []
                    bucket_len = 0
                chunks.extend(self._fixed_chunks(block))
                continue
            if bucket_len + block_len > self.char_limit and bucket:
                chunks.append("\n".join(bucket))
                bucket = [block]
                bucket_len = block_len
            else:
                bucket.append(block)
                bucket_len += block_len
        if bucket:
            chunks.append("\n".join(bucket))
        return chunks

    def _split_by_headers(self, text: str):
        lines = text.splitlines()
        blocks = []
        current = []
        header_pattern = re.compile(r"^#{1,3}\s+.+")
        for line in lines:
            if header_pattern.match(line):
                if current:
                    blocks.append("\n".join(current).strip())
                    current = []
            current.append(line)
        if current:
            blocks.append("\n".join(current).strip())
        return [b for b in blocks if b]

    def _fixed_chunks(self, text: str):
        return [text[i:i + self.char_limit] for i in range(0, len(text), self.char_limit)]

# Example usage with existing Crawl4AI scrape result
progress = ProgressTracker()
extract_node = NotebookCrawl4AIExtractNode(char_limit=12000)
state_after_extract = await extract_node.execute(result, progress)
print("Extracted records:", len(state_after_extract.get("extracted_items", [])))
```
